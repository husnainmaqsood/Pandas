```python
import pandas as pd
```

```python
df = pd.read_excel('EmployeeData.xlsx')
```

```python
df.head()
```

```python
df['Gender'].unique()
```

```python
gender_map = {'m':'Male','f':'Female'}
df['Gender'] = df['Gender'].map(gender_map)
```

```python
df['Gender']
```

```python
df['Jobcat']
```

```python
job_cat_map = {1:'Low',2:'Medium',3:'High'}
df['Jobcat'] = df['Jobcat'].map(job_cat_map)
```

```python
df.head()
```

```python
df.pivot_table(index= 'Gender',values = ['Minority', 'Prev Exp', 'Salary'],aggfunc='sum')
```

```python
df.pivot_table(index= ['Gender','Jobcat'],values = ['Minority', 'Prev Exp', 'Salary'],aggfunc='mean')
```

```python
df.pivot_table(index= ['Gender','Jobcat'],values =['Minority','Prev Exp','Salary'],aggfunc='sum',margins=True,margins_name='Total')
```

```python
df.head()
```

```python
df.pivot_table(index='Jobcat',values='Salary',columns='Gender',aggfunc='count',fill_value=0,margins=True,margins_name='Total')
```

```python
df.pivot_table(columns=['Gender','Jobcat'],values = ['Minority','Prev Exp','Salary'],aggfunc='sum')
```
